## Create Training/Test Sets

First, we're going to define some helper methods, then we'll generate the sets that we need to use

```python
def _loadAndResize(img_path, w, h):
    """
    Load the image at the path with a different given width and height
    """
    return image.load_img(img_path, target_size = (w,h))

# testing this thing
sample = _loadAndResize(sample_img_path, 64, 64)
plt.imshow(sample)
```

```python
def _generateLabel(path_string):
    """
    Labels img as 1 if it's a cat, 0 if it's anything else
    """
    if "cat" in path_string:
        return 1
    
    return 0

def _createDataset(train_dir_path):
    """
    creates X_train (and normalizes) and Y_train. X_train is list of images and Y_train is list of labels
    """
    train_img_path = os.listdir(train_dir_path)
    
    dir_list_length = len(train_img_path)
    
    X_train = np.zeros([25000, 64, 64, 3])
    Y_train = np.zeros([25000, 1])
    
    for index in range(dir_list_length):
        # create full path string to image
        full_path = TRAIN_IMG_DIR + "/" + str(train_img_path[index])
        
        # load the image and normalize
        img_array = np.asarray(_loadAndResize(full_path, 64, 64)) / 255
        
        # add image data to X_train and label data to Y_train
        X_train[index] = img_array
        Y_train[index] = _generateLabel(full_path)
        
        if index % 1000 == 0:
            print("Finished index " + str(index))
    
    print("Completed loading image dataset!")

    return X_train, Y_train
```

```python
X_train, Y_train = _createDataset(TRAIN_IMG_DIR)
```

```python
print("This image is a " + str(Y_train[23000]))
print("X_train shape: " + str(X_train.shape))
print("Y_train shape: " + str(Y_train.shape))
plt.imshow(X_train[23000])
```

## Model

Now we're going to make our model using Keras (with a Tensorflow backend). Nothing super special this time around, I'll be using the model based on the HappyHouse project from Andrew Ng's Deep Learning Specialization in Coursera


The model is very naive. It goes:

Input -> Convolutional Layer -> Max Pool -> Flatten -> Output

```python
def isCatModel(input_shape):
    X_input = Input(input_shape)
    
    X = ZeroPadding2D((3,3))(X_input)
    
    X = Conv2D(32, (7,7), strides = (1,1), name = 'conv0')(X)
    X = BatchNormalization(axis = 3, name = 'bn0')(X)
    X = Activation('relu')(X)
    
    X = MaxPooling2D((2,2), name = 'max_pool')(X)
    
    X = Flatten()(X)
    
    X = Dense(1, activation = 'sigmoid', name = 'fc')(X)
    
    model = Model(inputs = X_input, outputs = X, name = 'isCatModel')
    
    return model
```

Create the model 


```python
theModel = isCatModel(X_train.shape[1:])
```

Compile the model

```python
theModel.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])
```

Train the model


```python
theModel.fit(x = X_train[1:], y = Y_train[1:], epochs = 10, batch_size = 100)
```

Save the model so we can do stuff with it later
